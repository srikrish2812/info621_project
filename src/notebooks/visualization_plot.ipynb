```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plot (a)
error_types = [
    'Calculation Errors',
    'Semantic Misunderstandings',
    'Step-Missing Errors',
    'Units Misinterpretation',
    'Contextual Misunderstanding',
    'Potential Patterns in the Data',
    'Answer Formatting Errors'
]

# Data for the error counts from the two models
error_counts_model_a = [447, 643, 108, 35, 43, 23, 396]
error_counts_model_b = [422, 500, 72, 26, 33, 19, 270]

# Create the figure for side-by-side plots
fig, ax = plt.subplots(figsize=(12, 6))

# Set positions for the bars
bar_width = 0.35
index = np.arange(len(error_types))

# Plot the bars for Model A and Model B side by side
bars_model_a = ax.bar(index, error_counts_model_a, bar_width, label='Pre-trained gemma-3-1b-it-bnb-4bit model', color='#3498db', edgecolor='black')
bars_model_b = ax.bar(index + bar_width, error_counts_model_b, bar_width, label='Fine-tuned gemma-3-1b-it-bnb-4bit model', color='#2ecc71', edgecolor='black')

# Add labels and title
ax.set_xlabel('Error Type', fontsize=14)
ax.set_ylabel('Number of Samples', fontsize=14)
#ax.set_title('Comparison of Error Analysis for LLMs in Math Word Problems', fontsize=16)
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(error_types, fontsize=12, rotation=45, ha='right')

# Adding the data labels on top of the bars
for bar in bars_model_a:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 10, str(int(yval)), ha='center', va='bottom', fontsize=12, fontweight='bold')

for bar in bars_model_b:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 10, str(int(yval)), ha='center', va='bottom', fontsize=12, fontweight='bold')

# Add a legend
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
