```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plot
error_types = [
    'Calculation',
    'Semantic\nMisunderstandings',
    'Step-Missing',
    'Units\nMisinterpretation',
    'Contextual\nMisunderstanding',
    'Potential Patterns\nin the Data',
    'Answer Formatting'
]
error_counts = [447, 643, 108, 35, 43, 23, 396]

plt.style.use('ggplot')

# Create horizontal bar positions
y_pos = np.arange(len(error_types))

# Create a colormap
cmap = plt.get_cmap('viridis')
colors = cmap(np.linspace(0.2, 0.8, len(error_types)))

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6), dpi=100)
bars = plt.barh(y_pos, error_counts, color=colors, edgecolor='black')

# Invert y-axis so the largest bar is on top
plt.gca().invert_yaxis()

# Labels and title
plt.xlabel('Number of Samples', fontsize=14)
plt.ylabel('Error Type', fontsize=14)
plt.yticks(y_pos, error_types, fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```
